```python
def make_con_plot(df,title="",subtitle="",lim=100,ntrips=True):
    
    condf = make_con_df(df)
    
    start = 0.0
    stop = 1.0
    number_of_lines = len(condf)
    cm_subsection = linspace(start, stop, number_of_lines) 
    colors = [ cm.winter(x) for x in cm_subsection ]
    
    
    plot = mobi.geomobi.GeoPlot()
    plot.addgeo('/home/msj/shapes/bikeways.shp',facecolor="none",edgecolor='gray',zorder=95,alpha=0.1)
    plot.addgeo('/home/msj/shapes/shoreline2002.shp',facecolor='#ffffff',zorder=1)    
    
    plot.ax.text(plot.right,plot.top-500,title,color=plot.fg_color,size=18,alpha=0.8,horizontalalignment='right')
    plot.ax.text(plot.right,plot.top-1000,subtitle,color=plot.fg_color,size=18,alpha=0.8,horizontalalignment='right')
    if ntrips:
        plot.ax.text(plot.right,plot.top-1500,"trips = {:,}".format(len(df)),color=plot.fg_color,size=12,alpha=0.8,horizontalalignment='right')
    
    #for i,route in condf2.sort_values('trips',ascending=False).iloc[:500].sort_values('trips',ascending=True).iterrows():
    for i,route in condf.sort_values('trips',ascending=False).iloc[:lim].iterrows():
        [y1,x1] = route['start coords'] 
        [y2,x2] = route['stop coords']
        trips = route['trips']
        maxtrip = condf['trips'].max()
        #plot.ax.arrow(x1,y1,x2-x1,y2-y1,transform=ccrs.PlateCarree(),length_includes_head=False,alpha=0.1,width=trips/100000)
        line, = plot.ax.plot([x1,x2],[y1,y2],transform=ccrs.PlateCarree(),alpha=trips/maxtrip,linewidth=10*trips/maxtrip,color=plot.fg_color2) 
        #line, = plot.ax.plot([x1,x2],[y1,y2],transform=ccrs.PlateCarree(),alpha=0.5,linewidth=10*trips/maxtrip,color=plot.colors[1]) 

        line.set_solid_capstyle('round')
    return plot
```

```python
import matplotlib.dates as mdates
def plot_yoy(df,text=""):
    thdf = df.set_index('Departure').groupby(pd.Grouper(freq='H')).size()
    thdf = pd.DataFrame(thdf)
    thdf.columns = ['Trips']
    pdf = pd.pivot_table(thdf,index=thdf.index.dayofyear,columns=thdf.index.year,values='Trips',aggfunc='sum')
    pdf.index = pd.date_range('2017-01-01', periods=365,freq='D')
    pdf.columns.name = "Daily Trips"
    
    plot = mobi.plots.Plot()
    plot.ax.plot(pdf[2017],color=plot.fg_color)
    plot.ax.plot(pdf[2018],color=plot.fg_color2)
    plot.ax.xaxis.set_major_formatter(mdates.DateFormatter("%b"))
    plot.ax.legend()
    plot.ax.text(0.05,1,text,transform=ax.transAxes,color=plot.colors[1],size=18,alpha=0.8)
    
```

```python
#f,ax = plt.subplots()
plot_yoy(df,"All member types")
```

```python
plot_yoy(df[idx365all],"Annual member trips")
```

```python
plot_yoy(df[idx24],"24h pass trips")
```
